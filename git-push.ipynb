```python
import subprocess
import os

# Change to bot directory
os.chdir(r'c:\Users\rayya\OneDrive\Desktop\EmperorBot')
print('Working directory:', os.getcwd())
print('\n=== GIT OPERATIONS ===')

# Use the full path to git.exe
git_exe = r'C:\Program Files\Git\cmd\git.exe'

commands = [
    ('Check status', [git_exe, 'status', '--short']),
    ('Add all files', [git_exe, 'add', '.']),
    ('Commit changes', [git_exe, 'commit', '-m', 'Add welcome/farewell embeds']),
    ('Push to GitHub', [git_exe, 'push', 'origin', 'main']),
    ('Show latest commit', [git_exe, 'log', '-1', '--oneline']),
]

for step_name, cmd in commands:
    print(f'\n[{step_name}]')
    try:
        result = subprocess.run(cmd, capture_output=True, text=True, check=True)
        print(result.stdout if result.stdout else '(no output)')
        if result.stderr:
            print('STDERR:', result.stderr)
    except subprocess.CalledProcessError as e:
        print(f'ERROR: {e.returncode}')
        if e.stdout:
            print('STDOUT:', e.stdout)
        if e.stderr:
            print('STDERR:', e.stderr)
        break

print('\n=== DONE ===')
```
