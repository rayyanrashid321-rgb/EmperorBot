```python
import subprocess
import os
os.chdir(r'c:\Users\rayya\OneDrive\Desktop\EmperorBot')
print('cwd=', os.getcwd())
for cmd in [
    ['git', 'status', '--short'],
    ['git', 'diff', '--stat'],
    ['git', 'log', '-1', '--oneline'],
    ['git', 'branch', '--show-current'],
    ['git', 'add', '.'],
    ['git', 'commit', '-m', 'Add welcome/farewell embeds'],
    ['git', 'push', 'origin', 'main']
]:
    try:
        output = subprocess.check_output(cmd, stderr=subprocess.STDOUT, text=True)
    except subprocess.CalledProcessError as e:
        print('COMMAND FAILED:', ' '.join(cmd))
        print(e.output)
        break
    else:
        print('COMMAND:', ' '.join(cmd))
        print(output)
```
